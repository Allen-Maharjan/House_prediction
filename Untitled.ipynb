```python
#Feature engineering

#changing both numerical(data related to rating), date value to string
dataset['MSSubClass'] = dataset['MSSubClass'].astype(str)

dataset['OverallCond'] = dataset['OverallCond'].astype(str)

dataset['YrSold'] = dataset['YrSold'].astype(str)

dataset['MoSold'] = dataset['MoSold'].astype(str)


```

```python
#merging three column which given the surface area

dataset['TotalSF'] = dataset['1stFlrSF']+dataset['2ndFlrSF']+dataset['TotalBsmtSF']

dataset = dataset.drop(['1stFlrSF'],axis=1)
dataset = dataset.drop(['2ndFlrSF'],axis=1)
dataset = dataset.drop(['TotalBsmtSF'],axis=1)

dataset.head()
```

```python
#seprating categoracal feature and numerical feature

categorical_columns = [col for col in dataset.columns.values if dataset[col].dtype == 'object']

data_cat = dataset[categorical_columns]

data_num = dataset.drop(categorical_columns, axis=1)

data_cat.head()
```

```python
#checking the value of all numerical data if it is skewed or not

#univariate analysis
data_num.hist(figsize=(16,20), bins =25, xlabelsize =8, ylabelsize =8);
```

```python
#Checking the skewness of the data

from scipy.stats import skew

#calculating skew of data except Null 
data_num_skew = data_num.apply(lambda x : skew(x.dropna()))
data_num_skew = data_num_skew[data_num_skew > 0.75]


#using log transformation for making the data normally distributed (lop1p = log(1+x))
data_num[data_num_skew.index] = np.log1p(data_num[data_num_skew.index])
```

```python
data_num_skew
```

```python
data_num.hist(figsize=(16,20), bins = 50, xlabelsize =8, ylabelsize = 8);
```

```python
#Normalising the data (making the data range from 0-1)

#using the formula Xnorm = (X - Xmean)/(Xmax - Xmin)

data_num = (data_num - data_num.mean()) / (data_num.max() - data_num.min())

data_num.head()
```

```python
data_num.hist(figsize =(16,20), bins = 50, xlabelsize = 8, ylabelsize = 8);
```

<h3> Missing values for numerical data <h3>

```python
#missing value in numerical data 

#Missing value data analysis

Finding_null_values = dataset.isnull().sum()
Finding_null_values = Finding_null_values[Finding_null_values > 10]
Finding_null_values.sort_values(inplace= True)
Finding_null_values.plot.bar();
```

```python
#null values in percentage

total = dataset.isnull().sum().sort_values(ascending = False)
percent = (dataset.isnull().sum()/dataset.isnull().count()).sort_values(ascending = False)
missing_data = pd.concat([total,percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(15)
```

```python
#handling missing values

for col in data_num.columns.values:
    
    missing_values = data_num[col].isnull().sum()
    
    #droping columns if the missing values are greater than 227
    if missing_values > 227:
        data_num = data_num.drop(col, axis =1)
        
    else:
        
        data_num = data_num.fillna(data_num[col].median())
```

```python
data_num.head()
```

<h1> Missing values for categorical data <h1>
    

```python
#percentage of missing values

for col in data_cat.columns.values:
    missing_values = data_cat[col].isnull().sum()
    
    #droping column where missing value is greater than 50
    if missing_values > 50:
        print(f"droping column: {col}")
        data_cat.drop(col, axis=1)
    
```

```python
# dummy coding changing categorival values into dichotomous variable

data_cat_dummies = pd.get_dummies(data_cat, drop_first = True)

```

<h1> Linear Regression Modeling </h1>

```python
#splitting tha dataset into train and test
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(new_data, target_log, test_size = 0.30, random_state=0)
print(f"x_train {x_train.shape}")
print(f"x_test {x_test.shape}")
print(f"y_train {y_train.shape}")
print(f"y_test {y_test.shape}")
```

```python
#building linear regression model using stat model

import statsmodels.api as sm

#building using OLS

model1 = sm.OLS(y_train, x_train).fit()
```

```python
model1.summary()
```

```python
def rmse(prediction, target): #Root mean square error
    
    difference = prediction - target
    
    differences_squared = difference ** 2
    
    mean_of_differences_squared = differences_squared.mean()
    
    rmse_val = np.sqrt(mean_of_differences_squared)
    
    return rmse_val
```

```python
cols = ['Models', 'R-Squared Value', 'Adj_R-Squared Value', 'RMSE']

models_report = pd.DataFrame(columns = cols)

predictions1 = model1.predict(x_test)
```

```python
temp1 = pd.Series({
    'Models': "Base Linear Regression Model",
    "R-Squared Value": model1.rsquared,
    "Adj_R-Squared Value": model1.rsquared_adj,
    "RMSE": rmse(predictions1, y_test)
})

model1_report = models_report.append(temp1, ignore_index= True)
model1_report
```

<h1> Multicollinearity </h1>


```python
#calculating variance inflation Factor (VIF)

column_names = x_train1.columns
vif_values = []
for i in np.arange(0,len(column_names)):
    xvars = list(column_names)
    yvars = xvars.pop(i)
    mod = sm.OLS(x_train1[yvars], (x_train1[xvars]))
    res = mod.fit()
    vif = 1/(1-res.rsquared)
    vif_values.append(vif)
    print(yvars ,round(vif,3))
```

```python
#removing variable having VIF above 100

to_keep = []
counter  = 0
for x in x_train1:
    if vif_values[counter] < 100:
        to_keep.append(x)
    counter += 1
x_train2 = x_train1[to_keep]
x_train2.head()
```

```python
#Again building the model after removing VIF higher than 100

model3 = sm.OLS(y_train, x_train2).fit()

model3.summary()
```

```python
#removing value from test_Data whose VIF is greater than 100

x_test2 = x_test1[to_keep]
x_test.head()
```

```python
#Again predicting

predictions3 = model3.predict(x_test2)

tmp3 = pd.Series({
    'Models': 'Linear Regression Model with Constant',
    "R-Squared Value" : model3.rsquared,
    "Adj_R-Squared Value": model3.rsquared_adj,
    "RMSE": rmse(predictions3, y_test1)
})

model3_report = models_report.append(tmp3, ignore_index = True)
model3_report
```
